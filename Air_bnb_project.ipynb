## üßπ Step 4: Data Cleaning


```python
# No. of missing values in each column

data.isnull().sum()

# drop the missing value

data.dropna(inplace=True)

# watching missing value again

data.isnull().sum()

# checking the shape

data.shape
```

```python
# showing all the duplicate

data[data.duplicated()]
```

```python
# Deleating all duplicate

data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# Dealing with data inconsistency

# analysing all the data type first

data.dtypes
```

```python
# Changing id type from int to object

data["id"] = data["id"].astype("object")
data.dtypes

# changing host_id

data["host_id"] = data["host_id"].astype("object")
data.dtypes
```

**5. EDA**


**Univarient Analisis**

```python
# identifying outlier in price

df = data[data["price"]<1500]

sns.boxplot(data=df,x="price")
plt.savefig("price_boxplot.png")
```

```python
# price distribution

data["price"]

# Histogram for price distribution
plt.figure(figsize=(10,5))
sns.histplot(data=df,x="price",bins=100)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.savefig("price_distribution_histogram.png")
plt.show()
```

```python
df.dtypes
```

```python
# Availability frequency

plt.figure(figsize=(6,3))
sns.histplot(data=df,x="availability_365")
plt.title("Availability Frequency")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.savefig("availability_frequency_histogram.png")
plt.show()
```

```python
# comparing neighbourhood_group with avg price

df.groupby(by="neighbourhood_group")["price"].mean()
```

Feature Engineering

```python
# creating new clm, price per bed

df["price_per_bed"] = df["price"]/df["beds"]
df.head()
```

```python
# comparing neighbourhood_group with avg price per bed

df.groupby(by="neighbourhood_group")["price_per_bed"].mean()
```

```python
# Bargraph and graphical visualizion of neighbourhood_group with avg price per bed
mp = pd.DataFrame(df.groupby(by="neighbourhood_group")["price_per_bed"].mean())
plt.figure(figsize=(6,3))
sns.barplot(data=mp,x="neighbourhood_group",y="price_per_bed")
plt.title("price_per_bed")
plt.xlabel("neighbourhood_group")
plt.ylabel("Mean price")
plt.savefig("neighbourhood_group_price_per_bed_barplot.png")
plt.show()
```

**Bi Varient Analysis**

One variable dependency on another column

```python
df.columns
```

```python
# Price dependency on neighbourhood

sns.barplot(data=df,x="neighbourhood_group",y="price",hue="room_type")
plt.savefig("neighbourhood_group_price_barplot.png")
```

```python
# Number of review and price relation

plt.figure(figsize=(8,5))
plt.title("Number of review and price relation")
plt.xlabel("Number of review")
plt.ylabel("Price")
sns.scatterplot(data=df,x="number_of_reviews",y="price",hue="neighbourhood_group")
plt.savefig("reviews_price_scatterplot.png")
plt.show()
```

```python

df.dtypes
```

```python
# pairplot to understand all the possible outcomes

sns.pairplot(data=df,vars=["price","minimum_nights","number_of_reviews","availability_365"],hue="room_type")
plt.savefig("pairplot.png")
```

```python
# Geographical Distribution of airbnb
plt.figure(figsize=(10,7))
sns.scatterplot(data=df,x="longitude",y="latitude",hue="room_type")
plt.title("Geographical Distribution of airbnb")
plt.savefig("geographical_distribution_scatterplot.png")
plt.show()
```

```python
#Heatmap chart variable with other of numerical column
plt.figure(figsize=(10,7))
corr = df[["latitude","longitude","price","number_of_reviews","reviews_per_month","availability_365","beds"]].corr()
sns.heatmap(data=corr,annot=True)
plt.title("Heatmap chart variable with other of numerical column")
plt.savefig("correlation_heatmap.png")
plt.show()
```
